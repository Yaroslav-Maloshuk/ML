**case** - унікальний ідентифікатор кожного опосума в наборі. Просто порядковий номер. <br>
**site** - місце відлову опосума. Це може бути географічна точка (наприклад, ліс, узлісся, околиця поселення), код або назва району в Австралії чи Новій Гвінеї. <br>
**Pop** - популяція, до якої належить тварина. Наприклад, може позначати континентальну та острівну групу (що часто використовується у дослідженнях D. masculatus - австралійського опосума). <br>
**sex** – стать опосума: “M” для самця, “F” для самки. Часто використовується для дослідження статевого диморфізму (різниці у розмірах). <br>
**age** – вік, або ж ступінь дорослості (наприклад, дорослий/молодий). У деяких датасетах це може бути умовна категорія. <br>
**Морфометричні ознаки:**<br>
**hdlngth (head length)** – довжина голови, важлива для аналізу розмірів черепа. <br>
**skullw (skull width)** – ширина черепа. Часто корелює з масою тіла і використовується для видової класифікації. <br>
**totlngth (total length)** – загальна довжина опосума, включає тулуб і хвіст. <br>
**taill (tail length)** – довжина хвоста. У опосумів хвіст грає важливу роль, зокрема в маскуванні в середовищі, тому цей параметр показовий. <br>
**footlgth (foot length)** – довжина задньої лапи. Зазвичай використовується для оцінки рухових можливостей. <br>
**earconch (ear conch length)** – довжина вушної раковини. Цей показник допомагає у вивченні терморегуляції, бо у тварин з більшими вухами ефективніше теплообмін. <br>
**eye (eye diameter)** – діаметр або розмір ока. У нічних тварин, як опосуми, часто великі очі — це адаптація до умов низького освітлення. <br>
**chest** – ширина або обхват грудей. Показує загальний розвиток тіла. <br>
**belly** – ширина або обхват черевної частини тіла. Може відображати статеві, вікові або популяційні відмінності. <br>

```python
features = df[['hdlngth', 'skullw', 'totlngth', 'taill', 'footlgth', 'earconch', 'eye', 'chest', 'belly']]
features
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(features)
```

```python
wss = []
for i in range(1, 11):
  kmeans = KMeans(n_clusters=i, random_state=42)
  kmeans.fit(X_scaled)
  wss.append(kmeans.inertia_)

plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wss, marker='o', linestyle='--')
plt.title('Метод ліктя для визначення кількості кластерів')
plt.xlabel('Кількість кластерів')
plt.ylabel('WSS (Within-cluster Sum of Squares)')
plt.grid(True)
plt.show()
```

```python
silhouette_scores = []
for i in range(2, 11):
  kmeans = KMeans(n_clusters=i, random_state=42)
  kmeans.fit(X_scaled)
  score = silhouette_score(X_scaled, kmeans.labels_)
  silhouette_scores.append(score)

plt.figure(figsize=(8, 6))
plt.plot(range(2, 11), silhouette_scores, marker='o', linestyle='--')
plt.title('Метод силуетів для визначення кількості кластерів')
plt.xlabel('Кількість кластерів')
plt.ylabel('Силуетний коефіцієнт')
plt.grid(True)
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], hue=clusters, palette='viridis', legend='full')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label='Centroids')

plt.title('KMeans Clustering of Features with 3 Clusters')
plt.xlabel('Feature 1 (Scaled)')
plt.ylabel('Feature 2 (Scaled)')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

```python
dbscan = DBSCAN(eps=1.5, min_samples=5)
clusters_db = dbscan.fit_predict(X_scaled)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=clusters_db, palette='tab10', legend='full')

plt.title('DBSCAN Clustering (PCA Reduced)')
plt.xlabel('Principal Component 1')
plt.xlabel('Principal Component 2')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

```python
kmeans = MiniBatchKMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], hue=clusters, palette='viridis', legend='full')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label='Centroids')

plt.title('KMeans Clustering of Features with 3 Clusters')
plt.xlabel('Feature 1 (Scaled)')
plt.ylabel('Feature 2 (Scaled)')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

```python
from sklearn.decomposition import PCA

import matplotlib.pyplot as plt
import seaborn as sns
from kneed import KneeLocator

from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
from scipy.cluster.hierarchy import dendrogram

from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
```

```python
X = df

X = StandardScaler().fit_transform(X)

pca = PCA(random_state=42).fit(X)
pve = pca.explained_variance_ratio_
```

```python
sns.set_theme()

kneedle = KneeLocator(
    x=range(1, len(pve) + 1),
    y=pve,
    curve='convex',
    direction='decreasing'
)

kneedle.plot_knee()
plt.title(f'Knee Point at {kneedle.elbow + 1}')
plt.xlabel('Number of components')
plt.ylabel('Explained Variance Ratio')
plt.grid(True)
plt.show()
```

```python
n_components = kneedle.elbow

ax = sns.lineplot(np.cumsum(pve))

ax.axvline(x=n_components, c='black', linestyle='--', linewidth=0.75)

ax.axhline(y=np.cumsum(pve)[n_components], c='black', linestyle='--', linewidth=0.75)

ax.set(xlabel='number of components', ylabel='cumulative explained variance')

plt.show()
```

```python
model_kmn = KMeans(random_state=42)

visualizer = KElbowVisualizer(
    model_kmn,
    k=(2, 10),
    timings=False)

visualizer.fit(X)
visualizer.show()
```

```python
def plot_dendrogram(model, **kwargs):
  counts = np.zeros(model.children_.shape[0])
  n_samples = len(model.labels_)

  for i, merge in enumerate(model.children_):
    current_count = 0
    for child_idx in merge:
      if child_idx < n_samples:
        current_count += 1
      else:
        current_count += counts[child_idx - n_samples]
    counts[i] = current_count

  linkage_matrix = np.column_stack(
      [model.children_, model.distances_, counts]
  ).astype(float)

  dendrogram(linkage_matrix, **kwargs)
```

```python
model_agg = AgglomerativeClustering(
    distance_threshold=0,
    n_clusters=None)

model_agg = model_agg.fit(X)
```

```python
plot_dendrogram(model_agg, truncate_mode='level', p=3)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Number of points in node')

plt.show()
```

```python
k_best = visualizer.elbow_value_

model_kmn = KMeans(n_clusters=k_best, random_state=42).fit(X)
model_agg = AgglomerativeClustering(n_clusters=k_best).fit(X)

labels_kmn = pd.Series(model_kmn.labels_, name='k-means')
labels_agg = pd.Series(model_agg.labels_, name='h-clust')
```

```python
pd.crosstab(df['Pop'], labels_kmn)
```

```python
pd.crosstab(df['Pop'], labels_agg)
```

```python
fig, axes = plt.subplots(ncols=2, nrows=1, figsize=(10, 5), sharey=True)

for i, s in enumerate([df['Pop'], labels_kmn]):
  ax = axes[i]

  sns.scatterplot(x=X[:, 0],
                  y=X[:, 1],
                  hue=s,
                  style=s,
                  edgecolor='black',
                  linewidth=0.5,
                  s=60,
                  palette='tab20',
                  legend=False,
                  ax=ax)

  ax.set(title=s.name)

plt.show()
```
